# Classification 
## Logistic Regression Model
- Logistic Regression is a classification algorithm, Used to predict Binary Outcomes(0 or 1,Yes or No,True or False,Success or Failure,Male or Female,Head or Tail)

## How does logistic regression works
- It measures the relationship between input(Independent variable) and target(Dependent variable) by estimating probabilities using Sigmoid Function.

## Sigmoid Function
- This Function is an 'S'- Shaped Curve that map the real valued number (between 0 to 1)
- It can converts any line into curve.
  - p=1/(1+e^(-y)
  - y=mx+c
    -> p=1/(1+e^(-mx-c))
    -> ln(p/(1-p))=mx+c

```python
df=pd.read_csv('bid.csv')
df
```

```python
X=df['Number'].values
X
```

```python
Y=df['Win_Loss'].values
Y
```

```python
n=len(X)
print(np.ndim(X))
X=X.reshape(n,1)
print(np.ndim(X))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(C=1.0,solver='lbfgs',multi_class='ovr')
```

```python
lr.fit(X,Y)
log_predict=lr.predict(X)
```

```python
log_predict
```

```python
print(lr.predict_proba([[200]]))
```

```python
plt.scatter(X,Y)
plt.plot(X,log_predict)
```
